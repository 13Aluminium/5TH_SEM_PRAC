```python
import numpy as np
import pandas as pd


```

```python
df_book=pd.read_csv("BL-Flickr-Images-Book.csv")
df_book.head(15)

```

```python
df_book.shape
```

```python
df_book.loc[0]
```

```python
to_drop=['Edition Statement','Corporate Author','Corporate',
'Contributors','Former owner','Engraver','Contributors','Issuance',
'type','Shelfmarks']
df_book.drop(columns=to_drop,inplace=True)
df_book.head(15)
```

```python
df_book['Identifier'].is_unique
```

```python
df_book.set_index('Identifier',inplace=True)
df_book.loc[206]
```

```python
df_book.dtypes.value_counts()
df_book.dtypes
```

```python
df_book['Date of Publication'].head(15)
```

```python
regex=r'^(\d{4})'
date_num=df_book['Date of Publication'].str.extract(regex,expand=False)
date_num.head(15)
```

```python
df_book['Date of Publication']=pd.to_numeric(date_num,downcast='integer')
df_book['Date of Publication'].dtype
```

```python
df_book['Date of Publication'].isnull().sum()
```

```python
df_book['Date of Publication']=df_book['Date of Publication'].isnull().sum()!=True
df_book.head(15)
```

```python
book_idx=[4157862,4159587]
df_book.loc[book_idx]
```

```python
df_book['Place of Publication']=np.where(df_book['Place of Publication'].str.contains('London'),'London',df_book['Place of Publication'].str.replace('-',' '))
book_idx=[216,4157862,4159587]
df_book.loc[book_idx]
```

```python
null_counts=df_book.isnull().sum()
print(null_counts)
```

```python
df_book['Author']=df_book['Author'].str.strip().str.title()
df_book['Author'].fillna('unknown',inplace=True)
```

```python
df_book['Publisher']=df_book['Publisher'].str.strip().str.title()
df_book['Publisher'].fillna(df_book["Place of Publication"],inplace=True)
```
